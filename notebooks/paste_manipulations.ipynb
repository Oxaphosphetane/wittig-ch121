```python
def run_applescript(script):
    process = subprocess.Popen(['osascript', '-'], stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = process.communicate(script.encode('utf-8'))
    if process.returncode != 0:
        raise Exception(f"AppleScript failed: {stderr.decode('utf-8')}")
    return stdout.decode('utf-8')

def paste_text_line_by_line(text):
    lines = text.split('\n')
    new_doc_script = '''
        tell application "ChemDraw"
            activate
            tell application "System Events"
                keystroke "n" using command down
            
    for line in lines:
        # Copy the line to the clipboard
        pyperclip.copy(line)
        
        # AppleScript to paste the clipboard content and press Enter in ChemDraw
        script = '''
        tell application "ChemDraw"
            activate
            tell application "System Events"
                keystroke "v" using command down
                key code 36 -- press Enter key
            end tell
        end tell
        '''
        run_applescript(script)
        time.sleep(0.5)  # Add a small delay between lines if needed
```
